```python
import pandas as pd
import numpy as np
from pyecharts import options as opts
from pyecharts.charts import Bar,Line,Map,Tab
from pyecharts.globals import ThemeType,SymbolType

#读取文件
df_sheet1 = pd.read_excel('Country_Estimates_2021.xlsx',sheet_name=1)
df_sheet2 = pd.read_excel('Country_Estimates_2021.xlsx',sheet_name=2)
df_sheet3 = pd.read_excel('Country_Estimates_2021.xlsx',sheet_name=3)
df_sheet4 = pd.read_excel('Country_Estimates_2021.xlsx',sheet_name=4)
```

# 任务 3 数据探索和数据预处理

```python
#查看数据整体信息
# df_sheet1.info()
```

```python
#处理缺失值和异常值
def Mvao(x):
    x = x.drop('Note',axis=1)
    x.replace("-",np.nan,inplace=True)
    x = x.dropna()
    return x

df_sheet1 = Mvao(df_sheet1)
df_sheet2 = Mvao(df_sheet2)
df_sheet3 = Mvao(df_sheet3)
df_sheet4 = Mvao(df_sheet4)
```

```python
#提取该方法下2000 年至 2020 年的数据
df_sheet1_Point_Estimate = df_sheet1[df_sheet1['Estimate'].isin(['Point Estimate'])]
df_sheet2_Point_Estimate = df_sheet2[df_sheet2['Estimate'].isin(['Point Estimate'])]
df_sheet3_Point_Estimate = df_sheet3[df_sheet3['Estimate'].isin(['Point Estimate'])]
df_sheet4_Point_Estimate = df_sheet4[df_sheet4['Estimate'].isin(['Point Estimate'])]

# df_sheet1_Point_Estimate
```

# 任务 4 总体情况可视化分析

```python
Point_Estimate1 = df_sheet1_Point_Estimate.iloc[:,5:]
Point_Estimate2 = df_sheet2_Point_Estimate.iloc[:,5:]
Point_Estimate3 = df_sheet3_Point_Estimate.iloc[:,5:]
Point_Estimate4 = df_sheet4_Point_Estimate.iloc[:,5:]

#全世界发育不良平均比例
Point_Estimate1=Point_Estimate1.astype(float)
Point_Estimate_mean1 = Point_Estimate1.agg('mean',axis=0).round(2)

#全世界发育不良平均人数
Point_Estimate2=Point_Estimate2.astype(float)
Point_Estimate_mean2 = Point_Estimate2.agg('mean',axis=0).round(0)

#超重平均比例
Point_Estimate3=Point_Estimate3.astype(float)
Point_Estimate_mean3 =  Point_Estimate3.agg('mean',axis=0).round(2)

#超重平均人数
Point_Estimate4=Point_Estimate4.astype(float)
Point_Estimate_mean4 =  Point_Estimate4.agg('mean',axis=0).round(0)
```

```python
year = ['2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010',
        '2011','2012','2013','2014','2015','2016','2017','2018','2019','2020']

bar1 = (
    Bar(init_opts=opts.InitOpts(bg_color="pink")) 
    .add_xaxis(year)
    .add_yaxis("发育不良平均人数", Point_Estimate_mean2.values.tolist())
    .add_yaxis("超重平均人数", Point_Estimate_mean4.values.tolist())
    .extend_axis(yaxis=opts.AxisOpts(type_="value", position="right",min_=0,max_=30,axislabel_opts=opts.LabelOpts(formatter="{value} %") ))
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    .set_global_opts(
        title_opts=opts.TitleOpts(title="二十年儿童营养不良的变动趋势以及对比 2000 年与2020 年儿童营养不良的变化水平",pos_left=100),
        legend_opts=opts.LegendOpts(pos_top = 30, pos_right = 200)
    )
)

line = (
       Line()
       .add_xaxis(year)
       .add_yaxis("发育不良平均比例", Point_Estimate_mean1.values.tolist(),yaxis_index=1,areastyle_opts=opts.AreaStyleOpts(opacity=0.5))
       .add_yaxis('超重平均比例',Point_Estimate_mean3.values.tolist(),yaxis_index=1,areastyle_opts=opts.AreaStyleOpts(opacity=0.5))
       .set_series_opts(label_opts=opts.LabelOpts(is_show=False))

)
bar1.overlap(line)
bar1.render_notebook()
```

跟踪近二十年儿童营养不良的变动趋势以及对比 2000 年与2020 年儿童营养不良的变化水平，根据可视化结果进行分析。
